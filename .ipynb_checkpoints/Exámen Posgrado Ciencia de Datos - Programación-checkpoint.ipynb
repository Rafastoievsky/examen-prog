```python
import numpy as np
```

```python
covidOrig["Decesos"] =  covidOrig["FECHA_DEF"]
```

```python
covidOrig.loc[covidOrig['FECHA_DEF'] == '9999-99-99', 'Decesos'] = np.nan
```

```python
covidOrig
```

```python
tabla1Test = covidOrig.query('ENTIDAD_UM == 26 and RESULTADO == 1')[['FECHA_ACTUALIZACION','FECHA_SINTOMAS', 'ENTIDAD_UM','RESULTADO','Decesos']].groupby(['FECHA_SINTOMAS', 'ENTIDAD_UM', 'FECHA_ACTUALIZACION']).count()
```

```python
tabla1E1 = covidOrig.query('ENTIDAD_RES == 26 and RESULTADO == 1')[['FECHA_SINTOMAS', 'FECHA_INGRESO','FECHA_DEF','ENTIDAD_RES','RESULTADO']]
```

```python
#Diccionario de nombre de estados
estados = {8:'Chichuahua', 19:'Nuevo León', 21:'Puebla', 26:'Sonora'}
```

```python
tabla2E1 = covidOrig.query('ENTIDAD_UM == [8,19,21,26] and TIPO_PACIENTE == 2')[['ENTIDAD_UM', 'TIPO_PACIENTE']].groupby('ENTIDAD_UM').count()
```

```python
tabla2E1["estados"] = tabla2E1.index
```

```python
tabla2E1["estados"] = tabla2E1["estados"].map(estados)
```

```python
tabla2E1.set_index("estados", inplace=True)
```

```python
tabla2E1.rename(columns={'TIPO_PACIENTE':'Hospitalizados'}, inplace=True)
```

```python
tabla2E1
```

```python
Grafica1 = tabla2E1.plot(kind="bar",figsize=(15,5))
Grafica1.set_title("Casos de Hospitalizados en Chihuahua, Nuevo León, Puebla y Sonora")
```

```python
imagenPlot = Grafica1.get_figure()
```

```python
imagenPlot.savefig('grafico1.png')
```
